#REAL WORLD DATA

IMPORT DATA FROM WEBSITE

```
import pandas as pd
path = 'https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv'

df = pd.read_csv(path)
df.head()
```

```
df.shape
```

SHOW DATA TYPE

```
df.info()
```

```
sns.countplot(x = df.type)
plt.show()
```

```
sns.barplot(x = 'country', y= 'cases', data = df)

plt.show()
```

CLEAN UP THE X-AXIS, ROTATE IT. MAKE PLOT BIGGER

```
sns.set(rc={'figure.figsize': (40,20)})
sns.barplot(x = 'country', y= 'cases', ci=None, data = df)
plt.xticks(rotation = 90)
plt.show()
```

TREND BY TIME SERIES

```
sns.set(rc={'figure.figsize': (40,20)})
sns.barplot(x = 'date', y= 'cases', ci=None, data = df)
plt.xticks(rotation = 90)
plt.show()
```

FILTER TOP COUNTRIES

```
df1 =df[(df['country'] == 'US') |(df['country'] == 'Russia')| (df['country'] == 'Brazil') ]
df1.shape
```

```
df1.head()
```

```
sns.lineplot(x = 'date', y = 'cases', hue = 'country', ci=None, data= df1)
plt.xticks(rotation = 90)
plt.show()
```

```
sns.reset_orig()

sns.barplot(x = 'country', y = 'cases', hue = 'type', ci = None, data = df1)
plt.show()
```

PLACMENT OF COUNTRIES ON DIFFERENT GRAPHS, CHANGE OF DISPLAY

```
sns.relplot(x = 'date', y = 'cases', row = 'country', 
            height =5, 
            aspect = 5, 
            ci=None, kind = 'line', data = df1)
plt.show()
```
